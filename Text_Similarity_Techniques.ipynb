### **What Are Word Embeddings?**

Imagine you want to turn words into numbers so that a computer can understand and process them. Word embeddings are a way to do this by representing words as vectors, which are lists of numbers.

### **Why Use Word Embeddings?**

1. **Capturing Meaning:** Unlike simple one-hot vectors (which are just long lists of zeros with a single one), word embeddings can capture the meaning of words. For instance, in a good word embedding system, words like "cat" and "kitten" will have similar vectors because they are related.

2. **Efficiency:** Word embeddings are low-dimensional (meaning they use fewer numbers) compared to one-hot vectors, which are high-dimensional and sparse (mostly zeros).

### **Common Word Embedding Techniques:**

1. **Word2Vec:**
   - **What It Is:** A popular method for creating word embeddings.
   - **How It Works:** It represents words as vectors in such a way that words with similar meanings are close to each other in the vector space.

   **Two Main Architectures of Word2Vec:**
   - **Continuous Bag of Words (CBOW):**
     - **How It Works:** It tries to predict a word based on the words around it (the context). For example, given the sentence "The cat sits on the mat," CBOW would try to predict "sits" if it knows "The," "cat," "on," "the," and "mat."
   - **Skip-gram:**
     - **How It Works:** It does the opposite of CBOW. It uses a word to predict the surrounding words. So, if the word is "sits," it would try to predict "The," "cat," "on," "the," and "mat."



Sure! Let’s simplify Doc2Vec and how it helps with semantic similarity between sentences.

### **What Is Doc2Vec?**

Doc2Vec is an extension of Word2Vec that handles entire documents (or paragraphs) instead of just individual words. Just like Word2Vec creates vectors (lists of numbers) for words, Doc2Vec creates vectors for longer pieces of text, such as sentences or paragraphs.

### **How Doc2Vec Works:**

1. **Paragraph Vector and Word Vectors:**
   - **Word Vectors:** In Word2Vec, each word is turned into a vector. For example, the words "dog" and "cat" might have vectors that are close together because they are related.
   - **Paragraph Vector:** In Doc2Vec, each paragraph or document also gets a special vector, called the paragraph vector. This vector helps capture the meaning of the entire paragraph.

2. **PV-DM Model (Distributed Memory Model of Paragraph Vectors):**
   - **What It Is:** A specific way of using Doc2Vec where the paragraph vector is combined with word vectors to understand the meaning of a paragraph.
   - **How It Works:** Imagine a paragraph has a unique vector that stays the same across all sentences in that paragraph. This vector, along with the word vectors, helps in predicting the words in the paragraph.

3. **Training the Model:**
   - **Training Process:**
     - You start with random vectors for words and paragraphs.
     - Using a method called **stochastic gradient descent**, the model adjusts these vectors to better capture the meanings.
     - For each training example, a context (a small chunk of text) is sampled, and the model learns to predict words in this context, adjusting the vectors accordingly.

4. **Inference (Making Predictions with New Text):**
   - **When New Text Comes:** After the model is trained, it can handle new paragraphs.
   - **How It Works:** To find the paragraph vector for a new piece of text, the model uses what it learned during training to update the paragraph vector while keeping the word vectors fixed.

5. **Finding Similarity Between Sentences:**
   - **Comparing Paragraph Vectors:** For two sentences, the model finds their paragraph vectors and measures how similar these vectors are. If the vectors are close together, the sentences are considered similar.

## Doc2Vec

````python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Sample data
data = ["The movie is awesome. It was a good thriller",
		"We are learning NLP throughg GeeksforGeeks",
		"The baby learned to walk in the 5th month itself"]

# Tokenizing the data
tokenized_data = [word_tokenize(document.lower()) for document in data]

# Creating TaggedDocument objects
tagged_data = [TaggedDocument(words=words, tags=[str(idx)])
			for idx, words in enumerate(tokenized_data)]


# Training the Doc2Vec model
model = Doc2Vec(vector_size=100, window=2, min_count=1, workers=4, epochs=1000)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count,
			epochs=model.epochs)

# Infer vector for a new document
new_document = "The baby was laughing and palying"
print('Original Document:', new_document)

inferred_vector = model.infer_vector(word_tokenize(new_document.lower()))

# Find most similar documents
similar_documents = model.dv.most_similar(
	[inferred_vector], topn=len(model.dv))

# Print the most similar documents
for index, score in similar_documents:
	print(f"Document {index}: Similarity Score: {score}")
	print(f"Document Text: {data[int(index)]}")
	print()
````

## SBERT

Certainly! Let's break down SBERT (Sentence-BERT) and its different training objectives in simple terms.

### **What Is SBERT?**

SBERT is an extension of BERT (a popular language model) designed to create fixed-size embeddings (numerical representations) for sentences. Unlike BERT, which focuses on words and their contexts, SBERT generates embeddings for entire sentences or phrases, making it easier to compare and analyze them.

### **How SBERT Works:**

1. **BERT and Pooling:**
   - **Word Embeddings to Context Vectors:** SBERT starts by converting words in a sentence into embeddings (numerical vectors) using BERT. BERT then processes these embeddings to understand the context of the sentence.
   - **Pooling:** To get a fixed-size representation for the entire sentence, SBERT uses a pooling operation. The simplest and most effective pooling method found is mean pooling, where the embeddings of all words are averaged to form the sentence embedding.

### **Training Objectives of SBERT:**

SBERT uses three different methods to train the model and improve how well it generates sentence embeddings. Here’s a simple explanation of each:

1. **Classification Objective Function:**
   - **What It Is:** This approach trains SBERT to classify sentences into predefined categories based on labels.
   - **How It Works:**
     - **Siamese Network:** SBERT uses a Siamese network architecture for this. Imagine two identical BERT models running side by side. Both models share the same parameters, and their outputs are compared.
     - **How Training Happens:** Sentences are processed by both models to get their embeddings. The embeddings from each model, along with their difference, are used to predict a label using a softmax classifier.
     - **Objective Function:** The goal is to minimize the difference between the predicted labels and actual labels using cross-entropy loss.

     ```plaintext
     Objective Function = softmax(W^T(u, v, |u-v|))
     ```

     Here, `u` and `v` are the sentence embeddings from the two BERT models, `|u-v|` is the difference between these embeddings, and `W` is a weight matrix used for classification.

2. **Regression Objective Function:**
   - **What It Is:** This approach trains SBERT to predict a similarity score between pairs of sentences.
   - **How It Works:**
     - **Siamese Network:** Similar to the classification objective, this method also uses a Siamese network. However, instead of predicting discrete labels, it calculates how similar two sentences are.
     - **Cosine Similarity and Mean Squared Error (MSE):** After generating embeddings for both sentences, SBERT computes their cosine similarity. The model is trained to minimize the mean squared error between the predicted similarity score and the actual similarity score.

3. **Triplet Objective Function:**
   - **What It Is:** This approach trains SBERT to differentiate between similar and dissimilar sentences.
   - **How It Works:**
     - **Triplet Network:** In this setup, three sentences are processed simultaneously: an anchor sentence, a positive (similar) sentence, and a negative (dissimilar) sentence.
     - **Learning Goal:** The model learns to make the anchor sentence’s embedding closer to the positive sentence’s embedding while keeping it farther from the negative sentence’s embedding.
     - **Loss Function:** The model aims to minimize the distance between the anchor and positive embeddings and maximize the distance between the anchor and negative embeddings. The loss function used is:

     ```plaintext
     max(||sa − sp|| − ||sa − sn|| + ϵ, 0)
     ```

     Here, `sa`, `sp`, and `sn` are the embeddings for the anchor, positive, and negative sentences, respectively. `|| · ||` denotes the distance metric, and `ϵ` is a margin ensuring the positive example is sufficiently closer to the anchor than the negative example.



Sure, let’s walk through examples for each type of objective function used in SBERT (Sentence-BERT) training:

### **1. Classification Objective Function**

**Purpose:** To train SBERT to classify sentences into predefined categories based on labels.

**Example Scenario:**
Imagine you have a dataset of customer reviews and you want to classify them into different sentiment categories such as "positive," "neutral," and "negative."

**Training Data:**
- **Sentence 1:** "I love this product! It works great!" (Label: Positive)
- **Sentence 2:** "The product is okay, but it could be improved." (Label: Neutral)
- **Sentence 3:** "I am very disappointed with this purchase. It broke after one use." (Label: Negative)

**How It Works:**
1. **Create Sentence Embeddings:** For each sentence, the SBERT model generates embeddings using BERT.
2. **Siamese Network:** Two identical BERT models process the sentences to create embeddings.
3. **Softmax Classifier:** The embeddings, along with their differences, are fed into a softmax classifier with three nodes (one for each label: Positive, Neutral, Negative).
4. **Training Objective:** Minimize the difference between predicted and actual labels using cross-entropy loss.

**Output:** For a new review like "The item is decent but not amazing," the model predicts the most likely sentiment category based on the trained embeddings.

### **2. Regression Objective Function**

**Purpose:** To train SBERT to predict a continuous similarity score between pairs of sentences.

**Example Scenario:**
Suppose you want to measure the similarity between pairs of sentences in terms of their semantic meaning, for example, comparing sentence similarity in paraphrase detection.

**Training Data:**
- **Sentence Pair 1:** ("The weather is sunny today," "It is a bright and clear day.") with a similarity score of 0.9
- **Sentence Pair 2:** ("I need a ride to the airport," "The book was interesting.") with a similarity score of 0.1

**How It Works:**
1. **Create Sentence Embeddings:** Generate embeddings for both sentences in each pair.
2. **Siamese Network:** Process each sentence in a pair through identical BERT models to get embeddings.
3. **Cosine Similarity:** Calculate the cosine similarity between the two embeddings.
4. **Mean Squared Error (MSE):** Train the model to minimize the difference between the predicted similarity score and the actual score using MSE loss.

**Output:** For a new pair of sentences like ("I love traveling," "Exploring new places is my passion"), the model predicts a similarity score that reflects how closely they match in meaning.

### **3. Triplet Objective Function**

**Purpose:** To train SBERT to differentiate between similar and dissimilar sentences by minimizing the distance between an anchor and a positive sentence while maximizing the distance between the anchor and a negative sentence.

**Example Scenario:**
You want to train a model to recognize whether sentences are paraphrases or not.

**Training Data:**
- **Anchor Sentence:** "I am looking for a new phone."
- **Positive Sentence (similar):** "I need a new smartphone."
- **Negative Sentence (dissimilar):** "I am going to the grocery store."

**How It Works:**
1. **Create Sentence Embeddings:** Generate embeddings for the anchor, positive, and negative sentences.
2. **Triplet Network:** Use three identical BERT models (one for each sentence) to get their embeddings.
3. **Loss Function:** Train the model to minimize the distance between the anchor and positive embeddings (`sa` and `sp`), and maximize the distance between the anchor and negative embeddings (`sa` and `sn`).

   The loss function is:
   ```plaintext
   max(||sa − sp|| − ||sa − sn|| + ϵ, 0)
   ```
   Here, `ϵ` is a small margin to ensure that the positive example is closer to the anchor than the negative example by at least `ϵ`.

**Output:** After training, the model should produce embeddings where similar sentences (paraphrases) are close together, and dissimilar sentences are far apart in the vector space.

## USE - Universal Sentence Encoder

### What is Universal Sentence Encoder (USE)?

The Universal Sentence Encoder (USE) is a tool that turns sentences into numerical representations called **sentence embeddings**. These embeddings capture the meaning of the sentences in a way that allows you to compare and analyze them. USE can be used for various Natural Language Processing (NLP) tasks like text classification, similarity detection, and more.

### How Does USE Work?

USE has two main ways to encode sentences into embeddings:

#### 1. **Transformer-based Encoder**

- **What It Is**: A Transformer-based encoder is a type of deep learning model that uses attention mechanisms to understand the context of words in a sentence. The Transformer architecture used here has 6 layers, each with a self-attention module and a feed-forward network.
  
- **How It Works**:
  1. **Context-Aware Embeddings**: Each word in a sentence is transformed into a vector (numerical representation) that captures its meaning in the context of the other words.
  2. **Aggregation**: These vectors are then combined (added together) and adjusted by the length of the sentence to create a single 512-dimensional vector. This vector represents the entire sentence.

#### 2. **Deep Averaging Network (DAN)**

- **What It Is**: A simpler approach where the sentence is represented by averaging the embeddings of its words and bi-grams (pairs of consecutive words).
  
- **How It Works**:
  1. **Average Embeddings**: The embeddings for each word and bi-gram in the sentence are averaged.
  2. **Feed-Forward Network**: This average is then processed through a 4-layer neural network to produce a 512-dimensional vector that represents the sentence.

### Training the USE

The USE is trained using various techniques and tasks to learn how to encode sentences effectively:

1. **Tokenization**: First, sentences are broken down into tokens (words or sub-words) and converted to lowercase.
2. **Encoding**: Depending on whether you use the Transformer or DAN approach, the tokens are converted into a 512-dimensional vector.
3. **Training**: The model is trained using these embeddings to improve how well it captures sentence meanings. It is trained on both supervised tasks (like classification) and unsupervised tasks (like predicting sentence similarity).

````python
# Load the embeddings
data = np.load('/content/drive/MyDrive/LLM phd/USE_embeddings.npz')
test_vec = data['test']
sentences_vecs = data['sentences']

# Define a function to compute cosine similarity
def cosine_similarity(vec1, vec2):
    return 1 - distance.cosine(vec1, vec2)

# New test sentence
new_test = ["I enjoyed the film a lot"]

sentences = [
    ["The movie is nice and It was a good thriller"],
    ["We are learning NLP through GeeksforGeeks"],
    ["The baby learned to walk in the 5th month itself"]
]

# Compute the new test sentence embedding
new_test_vec = embed(new_test)[0]

# Compute similarity scores for the new test sentence
for i, sent_vec in enumerate(sentences_vecs):
    similarity_score = cosine_similarity(new_test_vec, sent_vec)
    print(f'\nFor sentence {i+1}: {sentences[i]}')
    print(f'Similarity Score = {similarity_score}')
````
